##  *DRILL - Descriptive Statistics and Normality

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

binomial_1 = np.random.binomial(30, .5, 100)

np.mean(binomial_1) # 14.93
np.std(binomial_1, ddof = 1) #2.8858939409451345
plt.hist(binomial_1)
plt.axvline(binomial_1.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(binomial_1.mean() + binomial_1.std(), color='r', linestyle='dashed', linewidth=2)

plt.axvline(binomial_1.mean()-binomial_1.std(), color='r', linestyle='dashed', linewidth=2) 


plt.show()
```

```python
exponential_2 = np.random.exponential(2, 100)
np.mean(exponential_2) # 2.0882642058674259
np.std(exponential_2, ddof = 1) # 2.1357088667439652
plt.hist(exponential_2)
plt.axvline(exponential_2.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(exponential_2.mean() + exponential_2.std(), color='r', linestyle='dashed', linewidth=2)

plt.axvline(exponential_2.mean()-exponential_2.std(), color='r', linestyle='dashed', linewidth=2) 


plt.show()
```

```python
poisson_3 = np.random.poisson(50, 100)
np.mean(poisson_3) # 49.689999999999998
np.std(poisson_3, ddof = 1) # 6.9814979145820226
plt.hist(poisson_3)
plt.axvline(poisson_3.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(poisson_3.mean() + poisson_3.std(), color='r', linestyle='dashed', linewidth=2)

plt.axvline(poisson_3.mean()-poisson_3.std(), color='r', linestyle='dashed', linewidth=2) 

plt.show()
```

```python
triangular_4 = np.random.triangular(3,5,8, 100)
np.mean(triangular_4) # 5.4516761821570876
np.std(triangular_4, ddof = 1) # 1.0105388813615968
plt.hist(triangular_4)
plt.axvline(triangular_4.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(triangular_4.mean() + triangular_4.std(), color='r', linestyle='dashed', linewidth=2)

plt.axvline(triangular_4.mean()-triangular_4.std(), color='r', linestyle='dashed', linewidth=2) 

plt.show()
```

```python
uniform_5 = np.random.uniform(2, 7, 100)
np.mean(uniform_5) # 4.4332902828137888
np.std(uniform_5, ddof = 1) # 1.4130183377062122
plt.hist(uniform_5)
plt.axvline(uniform_5.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(uniform_5.mean() + uniform_5.std(), color='r', linestyle='dashed', linewidth=2)

plt.axvline(uniform_5.mean()-uniform_5.std(), color='r', linestyle='dashed', linewidth=2) 

plt.show()
```

```python
rayleigh_6 = np.random.rayleigh(2, 100)
np.mean(rayleigh_6) # 2.3655254881280259
np.std(rayleigh_6, ddof = 1) #1.4440298479494462
plt.hist(rayleigh_6)
plt.axvline(rayleigh_6.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(rayleigh_6.mean() + rayleigh_6.std(), color='r', linestyle='dashed', linewidth=2)

plt.axvline(rayleigh_6.mean()-rayleigh_6.std(), color='r', linestyle='dashed', linewidth=2) 

plt.show()
```

Q: Evaluate whether the descriptive statistics provided useful information about
the variable. Can you identify any common characteristics of the distributions
that could be usefully described using the mean and/or standard deviation,
versus the ones that could not?

A: For all but the second and fifth graph, the exponential one and uniform graph. The bulk of the data points in the other graphs all fall around the center of the standard deviation lines. The uniform graph was interesting, the high amount of lows and highs balanced each other and kept the mean in the center. Due to the large number of highs and lows in data points, one might not  want to use mean or standard deviation to measure central tendency for a uniform graph. The exponential graph had the majority of the data points on the left side of the mean line.

```python
norm_1 = np.random.normal(5, .5, 100)
norm_2 = np.random.normal(10, 1, 100)

super_norm = norm_1 + norm_2

mean = np.mean(super_norm) # 14.996347
sd = np.std(super_norm, ddof = 1) # 1.371065

plt.hist(super_norm)
plt.axvline(mean, color='b', linestyle='solid', linewidth=2)

plt.axvline(mean + sd, color='red', linestyle='dashed', linewidth=2)

plt.axvline(mean - sd, color='red', linestyle='dashed', linewidth=2)

plt.show()
```

Q: Evaluate the descriptive statistics against the data.

A: As one might expect originating from normalized data, the data looks good. The standard deviation is fairly low and the bulk of the data is in center of the lines as one would hope to see.
