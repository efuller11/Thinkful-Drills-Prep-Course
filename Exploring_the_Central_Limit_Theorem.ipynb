```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
%matplotlib inline
```

 <font color= red>1</font>

```python
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2')

#lets make standard deviation and mean lines
plt.axvline(pop1.mean(), color='blue', linestyle='solid', linewidth=2)
plt.axvline(pop1.mean() + pop1.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(pop1.mean() - pop1.std(), color='r', linestyle='dashed', linewidth=2)

plt.axvline(pop2.mean(), color='m', linestyle='solid', linewidth=2)
plt.axvline(pop2.mean() + pop2.std(), color='g', linestyle='dashed', linewidth=2)
plt.axvline(pop2.mean() - pop2.std(), color='g', linestyle='dashed', linewidth=2)
plt.legend(loc='upper right') 
plt.show()
```

    This is a reference for a 100 sample size, I chose to add lines to signify standard deviation 
    of both populations with dashed lines and the two means with a solid line. 

```python
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 

plt.axvline(sample1.mean(), color='blue', linestyle='solid', linewidth=2)
plt.axvline(sample1.mean() + sample1.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(sample1.mean() - sample1.std(), color='r', linestyle='dashed', linewidth=2)

plt.axvline(sample2.mean(), color='m', linestyle='solid', linewidth=2)
plt.axvline(sample2.mean() + sample2.std(), color='g', linestyle='dashed', linewidth=2)
plt.axvline(sample2.mean() - sample2.std(), color='g', linestyle='dashed', linewidth=2)

plt.legend(loc='upper right') 
plt.show()
```

#### Increase the size of your samples from 100 to 1000, then calculate the means and standard deviations for your new samples and create histograms for each.

```python
sample1x20 = np.random.choice(pop1, 20, replace=True)
sample2x20 = np.random.choice(pop2, 20, replace=True)

plt.hist(sample1x20, alpha=0.5, label='sample 1(20)') 
plt.hist(sample2x20, alpha=0.5, label='sample 2(20)') 

plt.axvline(sample1x20.mean(), color='blue', linestyle='solid', linewidth=2)
plt.axvline(sample1x20.mean() + sample1x20.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(sample1x20.mean() - sample1x20.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(sample2x20.mean(), color='m', linestyle='solid', linewidth=2)
plt.axvline(sample2x20.mean() + sample2x20.std(), color='g', linestyle='dashed', linewidth=2)
plt.axvline(sample2x20.mean() - sample2x20.std(), color='g', linestyle='dashed', linewidth=2)

plt.legend(loc='upper right') 
plt.show()
```

#### Repeat this again, decreasing the size of your samples to 20. What values change, and what remain the same?
