```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

df = pd.read_csv('ks_projects.csv')

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)

plt.scatter(df['backers'], df['usd pledged'])
plt.xlabel('Money Pledged')
plt.ylabel('Backers')
plt.title('Number of Backers Compared to how Much Received')

plt.subplot(1, 2, 2)

plt.scatter(df['state'], df['usd_goal_real'])
plt.ylabel('USD Goal Amount')
plt.title('Goal versus success')
plt.tight_layout()

plt.show()
```

The first two scatter plot tables I chose to group up using subplot. First is a graph showing the money pledged and the amount of backers they had to support it. There were a few odd results but most seemed to fall within a reasonable amount of one another. The second shows the goals set and whether they succeeded, failed or something else. Some of the kickstarters that failed had some very lofty goals in mind when started their project and well, it did not work out so well for them. The successful kickstarters were the ones that had the lower more realistic goals. As I understand it, canceled and suspended generally should be viewed the same as failed in most circumstances, the creator just did not want the loss on their record.


```python
plt.figure(figsize=(9, 4))
num_of_categories = df.main_category.value_counts()
legend_categories = ['Film & Video', 'Music', 'Publishing', 'Games',
                    'Technology', 'Design', 'Art', 'Food']

plt.pie(num_of_categories[7:])
plt.legend(legend_categories)
plt.axis('equal')
plt.title('Most Popular Catagories')
plt.show()
```

Next up is my pie chart. This shows the breakdown in Kickstarter catagories. I chose the most popular eight, even among these it is easy to see that music and film & video are the most  popular among kickstarter creators. They cover nearly half of the kickstart market share in 2018. I was surprised at the results, as I have only used kickstarter a couple of times in the past for games. I did not realise music and video had such a large space in kickstarter until now.

```python
plt.hist(df.state)
plt.title('Amount of Kickstarters started and Whether They Were a Success')
plt.xlabel('State of the Kickstarters in 2018 so far')
plt.ylabel('Amount')
plt.show()
```

This last graph, a histograph reveals the not so pleasant truth on Kickstarter. At a glance, you will see that far more kickstarters fail than succeed. As mentioned above the canceled were likely going to fail and the creator cancelled it rather than let it fail.
